Cargar el modelo de análisis de sentimientos

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer, pipeline

# Cargar el modelo de análisis de sentimientos desde Hugging Face
sentiment_model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
sentiment_tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")

# Crear pipeline de análisis de sentimientos
sentiment_analysis = pipeline("sentiment-analysis", model=sentiment_model, tokenizer=sentiment_tokenizer)

# Función para obtener el sentimiento del texto
def obtener_sentimiento(texto):
    resultado = sentiment_analysis(texto)
    return resultado[0]['label'], resultado[0]['score']
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Cargar el modelo GPT-2 para el chatbot
gpt2_model = AutoModelForCausalLM.from_pretrained("gpt2")
gpt2_tokenizer = AutoTokenizer.from_pretrained("gpt2")

# Función para generar respuestas del chatbot
def generar_respuesta(texto_usuario):
    inputs = gpt2_tokenizer(texto_usuario, return_tensors="pt")
    output = gpt2_model.generate(inputs['input_ids'], max_length=50, num_return_sequences=1)
    respuesta = gpt2_tokenizer.decode(output[0], skip_special_tokens=True)
    return respuesta
```

Cargar el modelo Llama para el chatbot

Integrar el análisis de sentimientos con el chatbot

```python
# Función que ajusta la respuesta del chatbot según el sentimiento
def chatbot_con_sentimiento(texto_usuario):
    sentimiento, confianza = obtener_sentimiento(texto_usuario)

    # Generar respuesta del chatbot
    respuesta_chatbot = generar_respuesta(texto_usuario)

    # Ajustar respuesta según el sentimiento detectado
    if sentimiento == 'NEGATIVE':
        respuesta_ajustada = f"Lamento escuchar eso. {respuesta_chatbot}"
    elif sentimiento == 'POSITIVE':
        respuesta_ajustada = f"¡Me alegra que estés bien! {respuesta_chatbot}"
    else:
        respuesta_ajustada = respuesta_chatbot

    return respuesta_ajustada
```

Probar el chatbot

```python
# Probar el chatbot
while True:
    texto_usuario = input("Tú: ")
    if texto_usuario.lower() in ["salir", "exit"]:
        print("Chatbot: ¡Hasta luego!")
        break

    respuesta = chatbot_con_sentimiento(texto_usuario)
    print(f"Chatbot: {respuesta}")
```
